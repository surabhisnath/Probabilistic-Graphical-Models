# Monte Carlo Sampling

```python
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample simulation and plot
a = 5
num_points = 100
plt.figure(figsize=(a,a))
xx = np.linspace(0, a, 100)
yy = np.sqrt(a*a - xx*xx)
plt.fill_between(xx, 0, yy, color='pink')
x = np.random.uniform(0, a, num_points)
y = np.random.uniform(0, a, num_points)
plt.scatter(x, y, color='b')
```

```python
pival = 0
for a in range(1, 11):

    num_points = 1000

    x = np.random.uniform(0, a, num_points)
    y = np.random.uniform(0, a, num_points)

    quad = np.where(x*x + y*y - a*a < 0, 1, 0)

    num_quad = sum(quad)
    num_sq = len(quad)
    piby4 =  num_quad / num_sq
    pi = piby4 * 4
    
    pival += pi

print("The approximated value of pi is =", pival/10)
```
